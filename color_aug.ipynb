```python
from PIL import Image
import numpy as np
import random
import pylab
img = np.array(Image.open('cat.jpg'), dtype=np.float32)/255
pylab.imshow(img)
pylab.imshow(img)
```

```python
img[:,:,0] *= random.random()*0.2+0.9
img[:,:,1] *= random.random()*0.2+0.9
img[:,:,2] *= random.random()*0.2+0.9
pylab.imshow(img)
```
