```python
import os
import openai

# openai.api_key =
```

```python
message_list = [{
        "role": "user",
        "content": "Provide a list of organic chemicals in their IUPAC names\
        consisting of 10 or fewer carbon atoms. DO NOT MAKE LONG CARBON CHAINS THAT IS LONGER THAN 6 CARBONS. These compounds should be\
        easily recognizable for a 3rd-year chemistry student. Arrange\
        the list such that similar structures are together,\
        making sure that each chemical is only\
        differentiated from its adjacent neighbors\
        by substitution, addition, or elimination reactions. \
        \nYou should provide a chuck of info (about 200) at a time,\
        be expected to provide for about 2k items.\
        Only provide the IUPAC name and nothing else.. \
        Use only elements of C, O, S, H Cl, N, Br, F, and P\nDo Not number the list.\nFor instance:\nBenzene \nMethylbenzene\n1,2-Dimethylbenzene\n1,4-Dimethylbenzene\nPhenyl chloride\n..."
        }]
for i in range(10):
  response = openai.ChatCompletion.create(
    model="gpt-4",
    messages=message_list,
    stream = True
  )
  msg = ""
  for i in response:
    try:
      t = i["choices"][0]["delta"]["content"]
      print(t, end="")
      msg+=t
    except:
      pass

  message_list.append(
    {
      "role": "assistant",
      "content": msg
    }
         )

```

```python
for i in range(10):
  response = openai.ChatCompletion.create(
    model="gpt-4",
    messages=message_list,
    stream = True
  )
  msg = ""
  for i in response:
    try:
      t = i["choices"][0]["delta"]["content"]
      print(t, end="")
      msg+=t
    except:
      pass

  message_list.append(
    {
      "role": "assistant",
      "content": msg
    }
         )

```

```python
items = []
```

```python

for i in message_list:
  if i["role"] == "assistant":
    items.extend([(i[:-1] if i.endswith(",") else i) for i in i["content"].split("\n")])
```

```python
with open("items.txt","w") as f:
  f.write("\n".join(items))
```
